```python
import numpy as np
import pandas as pd
from numpy.random import randn
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
plt.style.use("ggplot")
```

```python
dataset = randn(100)
```

- ヒストグラムとPDFをまとめて描画

```python
sns.distplot(dataset)
```

- rugplotを追加する
- ヒストグラムを非表示にする

```python
sns.distplot(dataset, rug=True, hist=False)
```

- それぞれのプロット毎に必要な設定を引数で与えることもできる

```python
sns.distplot(dataset, bins=50,
            kde_kws={'color': 'indianred', 'label': 'KDE PLOT'},
            hist_kws={'color': 'blue', 'label': 'HISTGRAM'})
```

- Seriesと合わせて使う

```python
ser1 = pd.Series(dataset, name='My_DATA')
```

```python
ser1
```

```python
sns.distplot(ser1, bins=20)
```
