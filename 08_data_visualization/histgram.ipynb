```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
plt.style.use("ggplot")
```

```python
dataset1 = np.random.randn(100)
```

```python
plt.hist(dataset1, bins=20)
```

```python
dataset2 = np.random.randn(80)
```

```python
plt.hist(dataset2, bins=20, color="indianred")
```

normedで標準化すると、全体を積分した時に合計が１になる。

```python
plt.hist(dataset1, normed=True, bins=20)
```

標準化してやると、縦軸の数字が揃うので、グラフを重ねて書くことができる。

```python
plt.hist(dataset1, bins=20, normed=True, alpha=0.5)
plt.hist(dataset2, bins=20, normed=True, alpha=0.5)
```
